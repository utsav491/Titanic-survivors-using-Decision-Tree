```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import export_graphviz
from IPython.display import Image

df =pd.read_csv("C:\\Users\\utsav\\Desktop\\Titanic.csv" , index_col = 0)
```

```python
#Create a function that creates dummy variables  
def dummy(column):
    column = pd.get_dummies(df[column] , drop_first = True)
    return column
```

```python
df["survived"] = df["survived"].astype(str)
df["survived"].dtype
```

```python
column_list = ["pclass" , "sex"]
df_1 = dummy(column_list)
print(df_1.head(10))
```

```python
#Columns to be dropped
to_be_dropped = ["pclass",  "name",  "sex", "parch", "ticket", "fare", "cabin", "embarked", "boat", "body", "home.dest"]
df.drop(to_be_dropped, axis = 1 , inplace = True)
```

```python
#Getting the final concated Data Frame
df_titanic = pd.concat( [df, df_1], axis =1)
print(df_titanic.head())
```

```python
df_titanic =  df_titanic.iloc[:, 0:6]

#for i in
#d =list(o.columns) 
#print(d) 
#df[i] = df[i].astype("category")
p =(df_titanic.columns)
print(p)
for u in p:
    if u == 'age' or u =='survived':
        continue
    df_titanic[u] = df_titanic[u].astype('category')
```

```python
# Checking for Data Types and any missing values
df_titanic.info()
```

```python
#Repalcing the null values with the mean values
df_titanic['age'].fillna((df_titanic['age'].mean()), inplace=True)
df_titanic.isna().sum()
```

```python
#Creating X column
X = df_titanic.drop(["survived"], axis = 1)
print(X.shape)
print(X.columns)
print(X.head())
```

```python
#Creating y column
y = df_titanic["survived"]
print(y.shape)
```

```python
#Train and test split
from sklearn.model_selection import train_test_split
X_train, X_test ,y_train,y_test = train_test_split(X,y,test_size =0.3 ,random_state = 42) #staritfy = X["age"])
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 41)
```

```python
dt.fit(X_train, y_train)
export_graphviz(dt , out_file = 'dt.dot' , feature_names= ['age' , 'sibsp' ,'pclass_2','pclass_3','sex_male' ] , class_names= ['1','0'])
```

```python
Image('dt.png', width = 950 ,height = 950)
```

```python
"Calculating training set error"
from sklearn.metrics import accuracy_score

dt.fit(X_train, y_train)
training_set_prediction = dt.predict(X_train)
training_set_results  = accuracy_score(training_set_prediction , y_train)
print("The trainig set  result is {}".format(training_set_results * 100))
```

```python
" Calculating Test set Error"
dt.fit(X_train ,y_train)
test_set_prediction  = dt.predict(X_test)
test_set_accuracy = accuracy_score(test_set_prediction , y_test)
#print(test_set_accuracy)
print("The test set  result is {}".format(test_set_accuracy * 100))
```

```python

from sklearn.metrics import confusion_matrix
cm_train_set = confusion_matrix(training_set_prediction , y_train)
print(cm_train_set)
```

```python
print(cm_train_set.ravel())
```

```python
tp ,fp , fn , tn = cm_train_set.ravel()

print(tp, fn)
```

```python
tpr = (tp)/(tp +fn)
print(tpr)
print("in‐sample percent survivors correctly predicted: " , tpr * 100)
```

```python
tnr = tn / (tn+ fp)
#print(tnr)
print("in-sample percent fatalities correctly predicted (on training set) : " , tnr * 100)
```

```python
" Test Set "
cm_test_set = confusion_matrix(test_set_prediction , y_test)
print(cm_test_set)
```

```python
tp_test , fp_test , fn_test ,tn_test = cm_test_set.ravel()

print(tp_test , fp_test , fn_test ,tn_test)
```

```python
tpr_test = (tp_test)/(tp_test + fn_test)
#print(tpr_test)

print("out‐of‐sample percent survivors correctly predicted (on test set) :" , tpr_test* 100)
```

```python
tnr_test = tn_test / (tn_test+ fp_test)
#print(tnr_test)
print("out-of‐sample percent fatalities correctly predicted (on test set) " ,tnr_test * 100)
```
